1. Quais foram as alterações ano a ano no ranking dos primeiros 10 países. 

```python
consulta_rank = uniao.loc[uniao["Happiness Rank"] <=10]
consulta_rank = consulta_rank.pivot_table(columns = 'Year',index = 'Country',values ="Happiness Rank", fill_value="-")

consulta_rank
```

Considerações das mudanças no ranking ano a ano

* Austrália se manteve na décima posição, até que saiu do top 10 em 2019, cedendo lugar para a Áustria.
* Queda no ranking do Canadá, saindo do 5ª posição para a 9ª
* Ascensão no ranking da Finlândia, saindo da 6ª posição para a 1ª nos últimos dois anos.
* Queda gradativa da Suiça ano a ano, saindo da 1ª posição para a 6ª.
* Os outros países não apresentaram grandes variações, se mantendo estável ou subindo/descendo um ranking.

2. Considerando a média de todas as variáveis, quais foram as 3 variáveis que mais mudaram ano a ano.(Sem contar o score ou ranking)

```python
#Calculando o desvio padrão das médias das variáveis, depois de tê-las dividido ano a ano
consulta_mean = uniao.groupby(['Year']).mean().std().round(2)
consulta_mean
```

```python
consulta_final =  uniao.groupby(['Year']).agg('mean').T
consulta_final['STD'] = consulta_mean
consulta_final[2:].round(2)
```

Analisando o desvio padrão, podemos concluir que as variáveis que tiveram maior dispersão foram Family, Health (Life Expectancy) e Economy (GDP per Capita)

3. Quais foram os top 3 países que mais tiveram mudança positiva e negativa de GPD do ano 2016 para 2017

```python
uni = uniao.loc[uniao["Year"]==2016].groupby(['Country'])["Economy (GDP per Capita)"].sum() - uniao.loc[uniao["Year"]==2017].groupby(['Country'])["Economy (GDP per Capita)"].sum()
uni_smallest=pd.DataFrame(uni.nsmallest(3))
uni_largest=pd.DataFrame(uni.nlargest(3))
```

```python
#Top 3 países com maior mudança positiva
uni_largest
```

```python
#Top 3 países com maior mudança negativa
uni_smallest
```

4. Qual foi o país que mais subiu no ranking no período inteiro analisado.

```python
top=uniao.loc[uniao["Year"]==2015].groupby(['Country'])["Happiness Rank"].sum() - uniao.loc[uniao["Year"]==2019].groupby(['Country'])["Happiness Rank"].sum()
top=pd.DataFrame(top.nlargest(1))
top
```

```python
top_pais=uniao.loc[uniao["Country"]==top.index[0]].groupby(["Year"])["Happiness Rank"].sum()
top_pais=pd.DataFrame(top_pais)
top_pais=top_pais.T
top_pais["Delta"]=top.iloc[0,0]
```

```python
top_pais.rename(index={"Happiness Rank":top.index[0]})
```

O país que mais subiu posições no período analisado foi Benin, subindo 53 posições.

7. Duas Questões para analisar

```python
#Comportamento do Brasil

consulta_brasil = uniao.loc[uniao["Country"] == 'Brazil']
consulta_brasil
```

```python
#Correlação dos índices

consulta_corr = uniao.iloc[:,3:9].corr().round(3)
consulta_corr
```

```python
#Densidade por cada variável

for item in consulta_corr.columns:
 
    hist_data = uniao[item]
    hist_data.plot.hist(bins=40, color = ["lightblue"], figsize=(10,4),
                        edgecolor='black', lw=1.2, title="Densidade de Respostas - " + item)
    plot.xlabel(item)
    plot.ylabel("Frequência")
    plot.show()
```
