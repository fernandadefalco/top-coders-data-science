**1)** Um posto est√° vendendo combust√≠veis com a seguinte tabela de descontos: <br>

√Ålcool:<br>
 - at√© 20 litros, desconto de 3% por litro
 - acima de 20 litros, desconto de 5% por litro <br>
 
Gasolina:<br>
 - at√© 20 litros, desconto de 4% por litro
 - acima de 20 litros, desconto de 6% por litro. 
 
Escreva um algoritmo que leia o n√∫mero de litros vendidos, o tipo de combust√≠vel (codificado da seguinte forma: A-√°lcool, G-gasolina), calcule e imprima o valor a ser pago pelo cliente sabendo-se que o pre√ßo do litro da gasolina √© 2,50 reais o pre√ßo do litro do √°lcool √© 1,90 reais.

```python
preco_gas = 2.50
preco_alc = 1.90

desc_alc_ate_20 = 0.03
desc_alc_maior_20 = 0.05

desc_gas_ate_20 = 0.04
desc_gas_maior_20 = 0.06

litros = float(input("Digite a quantidade de litros vendidos: "))
combustivel = input("Informe o tipo de combust√≠vel (A-√°lcool ou G-gasolina): ").upper()

while combustivel != "A" and combustivel !="G":
    combustivel = input("Valor inv√°lido, informe o tipo de combust√≠vel (A-√°lcool ou G-gasolina): ").upper()

if litros > 20 and combustivel == "A":
    preco = (1-desc_alc_maior_20)*litros*preco_alc
elif litros <= 20 and combustivel == "A":
    preco = (1-desc_alc_ate_20)*litros*preco_alc
elif litros > 20 and combustivel == "G":
    preco = (1-desc_gas_maior_20)*litros*preco_gas
else:
    preco = (1-desc_gas_ate_20)*litros*preco_gas
    
print(f"Para abastecer {litros:.1f}L, o valor a ser pago √© R${preco:.2f}")
```

326 = 3 centenas, 2 dezenas e 6 unidades <br>
12 = 1 dezena e 2 unidades

**4)** Fa√ßa um programa que calcule as ra√≠zes de uma equa√ß√£o do segundo grau, na forma:

$ax^{2} + bx + c$ 

O programa dever√° pedir os valores de a, b e c, informando ao usu√°rio nas seguintes situa√ß√µes:
Se o usu√°rio informar o valor de A igual a zero, a equa√ß√£o n√£o √© do segundo grau e o programa n√£o deve fazer pedir os demais valores, sendo encerrado;
Se o delta calculado for negativo, a equa√ß√£o n√£o possui raizes reais. Informe ao usu√°rio e encerre o programa;
Se o delta calculado for igual a zero a equa√ß√£o possui apenas uma raiz real; informe-a ao usu√°rio;
Se o delta for positivo, a equa√ß√£o possui duas raiz reais; informe-as ao usu√°rio;

```python
a = float(input("Insira o valor de a: "))

#Se o usu√°rio informar o valor de A igual a zero, 
#a equa√ß√£o n√£o √© do segundo grau e o programa n√£o deve fazer pedir os demais valores, sendo encerrado
if a == 0:
    while a == 0:
        print("A equa√ß√£o n√£o √© de segundo grau, insira novamente: ")
        a = float(input("Insira o valor de a: "))
    
b = float(input("Insira o valor de b: "))
c = float(input("Insira o valor de c: "))

#Se o delta calculado for negativo, a equa√ß√£o n√£o possui raizes reais. 
delta = b**2 - 4*a*c

if delta < 0:
    print("A equa√ß√£o n√£o possui ra√≠zes reais.")

elif delta == 0:
    raiz = -b/(2*a)
    if raiz == -0:
        raiz = 0
    print(f"A equa√ß√£o s√≥ possui 1 raiz real: {raiz}")

else:
    raiz_1 = (-b + delta*(1/2))/(2*a)
    raiz_2 = (-b - delta*(1/2))/(2*a)
    print(f"A equa√ß√£o possui duas ra√≠zes reais: {raiz_1:.2f} e {raiz_2:.2f}")
```
